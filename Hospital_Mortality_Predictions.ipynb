```python
col = ['age','BMI','heart rate',
       'Systolic blood pressure', 'Diastolic blood pressure',
       'Respiratory rate', 'temperature', 'SP O2', 'Urine output',
       'hematocrit', 'RBC', 'MCH', 'MCHC', 'MCV', 'RDW', 'Leucocyte',
       'Platelets', 'Neutrophils', 'Basophils', 'Lymphocyte', 'PT', 'INR',
       'NT-proBNP', 'Creatine kinase', 'Creatinine', 'Urea nitrogen',
       'glucose', 'Blood potassium', 'Blood sodium', 'Blood calcium',
       'Chloride', 'Anion gap', 'Magnesium ion', 'PH', 'Bicarbonate',
       'Lactic acid', 'PCO2', 'EF']
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df[col] = SS.fit_transform(df[col])
df['gendera'].replace({1:0,2:1},inplace = True)
df
```

```python
X = df.iloc[:,1:]
Y = df.iloc[:,0:1]
```

Model Building

```python
from sklearn.model_selection import train_test_split
X1_train,X1_test,Y1_train,Y1_test = train_test_split(X,Y,test_size = 0.3, random_state = 42)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X1_train,Y1_train)
y_predlog = LR.predict(X1_test)
from sklearn.metrics import classification_report
print(classification_report(y_predlog,Y1_test))
```

```python
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN
SM = SMOTEENN()
X1,Y1 = SM.fit_resample(X,Y)
X_train,X_test,Y_train,Y_test=train_test_split(X1, Y1,test_size=0.3,random_state = 42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
LR.fit(X_train,Y_train)
y_predlog = LR.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_predlog,Y_test))
```

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score
K = StratifiedKFold(n_splits = 10)
result_log = cross_val_score(LR, X,Y, cv=K,n_jobs = -1)
result_log
```

```python
result_log
print("Stratified K Fold" ,np.mean(result_log).round(2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors' : [1,3,5,7,9,11]
}
Search_KNN = GridSearchCV(KNN, param_grid, cv=5,n_jobs = -1)
res_KNN = Search_KNN.fit(X,Y)
print(Search_KNN.best_params_)
print(Search_KNN.best_score_)
```

```python
KNN = KNeighborsClassifier(n_neighbors=5)
KNN.fit(X_train,Y_train)
y_pred_knn = KNN.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_pred_knn,Y_test))
```

```python
#Model Validation
K = StratifiedKFold(n_splits = 10)
result_KNN = cross_val_score(KNN, X,Y,cv = K)
print(result_KNN)
```

```python
print("Stratified for KNN is the ", np.mean(result_KNN).round(2))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
param_grid = {
    "criterion" : ["gini", "entropy"],
    "splitter" : ["best", "random"],
    "max_depth" : [2,3,5,7,9],
    "min_samples_leaf": [1,2,3,5,7,9],
    "min_samples_split": [2,5,7,9,10]
}
gd_D = GridSearchCV(DT,param_grid,cv= 5, n_jobs = -1)
result_D = gd_D.fit(X,Y)
print(result_D.best_score_)
print(result_D.best_params_)
```

```python
DT = DecisionTreeClassifier(criterion = "gini",splitter = "random", max_depth = 2,min_samples_leaf=1,min_samples_split=2 )
DT.fit(X_train,Y_train)
y_pred = DT.predict(X_test)
print(classification_report(y_pred,Y_test))
```

As we can see that predicting of the 0's and 1's are inconsistent we use the smoteen technique


```python
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN
SM = SMOTEENN()
X1,Y1 = SM.fit_resample(X,Y)
Xt_train,Xt_test,Yt_train,Yt_test=train_test_split(X1, Y1,test_size=0.3,random_state = 42)
print(Xt_train.shape,Xt_test.shape,Yt_train.shape,Yt_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
param_grid = {
    "criterion" : ["gini", "entropy"],
    "splitter" : ["best", "random"],
    "max_depth" : [2,3,5,7,9],
    "min_samples_leaf": [1,2,3,5,7,9],
    "min_samples_split": [2,5,7,9,10]
}
gd_D = GridSearchCV(DT,param_grid,cv= 5, n_jobs = -1)
result_D = gd_D.fit(Xt_train,Yt_train)
print(result_D.best_score_)
print(result_D.best_params_)
```

```python
DT = DecisionTreeClassifier(criterion = "gini",splitter = "best", max_depth = 9,min_samples_leaf=3,min_samples_split=2 )
DT.fit(Xt_train,Yt_train)
y_pred = DT.predict(Xt_test)
print(classification_report(y_pred,Yt_test))
```

```python
k = StratifiedKFold(n_splits = 10)
result_DT = cross_val_score(DT,X1,Y1,cv = k)
result_DT
```

```python
np.mean(result_DT).round(2)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFF = RandomForestClassifier()
param_grid = {
    "criterion": ["gini", "entropy"],
    'n_estimators': [100,200,300],
    'max_depth' : [2,3,4,5,7,9,11],
    'max_features': ["auto",'sqrt','log2','n_features'],

}

res = GridSearchCV(RFF,param_grid,cv = 5, n_jobs = -1)
res_Rff = res.fit(Xt_train,Yt_train)
print(res_Rff.best_params_)
print(res_Rff.best_score_)
```

```python
RFF = RandomForestClassifier(criterion = "gini", max_depth = 11, max_features = 'log2', n_estimators = 300)
RFF.fit(Xt_train,Yt_train)
y_predRff = RFF.predict(Xt_test)
print(classification_report(y_predRff,Yt_test))
```

```python
k= StratifiedKFold(n_splits=10)
result_rff = cross_val_score(RFF, Xt_train,Yt_train, cv = k)
```

```python
print(np.mean(result_rff).round(2))
```

XGBoost Classifier

```python
import xgboost as xgb
XGB = xgb.XGBClassifier()
param_grid = {
    "learning_rate": [0.1,0.2,0.001,0.002,0.3],
    "gamma" : [0.1,0.2,0.3,0.4,1,2,4,5,10],
    "max_depth": [3,4,5,7,9,11],
    "tree_method": ["auto","exact","approx","hisr"],
    }
resuult = GridSearchCV(XGB,param_grid,cv=5,n_jobs = -1, scoring='accuracy')
result_xb = resuult.fit(Xt_train,Yt_train)
print(result_xb.best_params_)
print(result_xb.best_score_)
```

```python
k = StratifiedKFold(n_splits=10)
result_xgb = cross_val_score(XGB, X1,Y1,cv=k)
print(result_xgb)
print(np.mean(result_xgb).round(2))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import GridSearchCV
BC = BaggingClassifier()
param_grid ={
    'n_estimators' : [200, 300, 400, 500],
    'max_samples' : [0.5, 0.6, 0.7, 0.8],
    'max_features' : [0.5, 0.6, 0.7, 0.8]
}
grid_result  = GridSearchCV(BC,param_grid,cv = 5,n_jobs = -1)
grid_result_BC = grid_result.fit(Xt_train,Yt_train)
print(grid_result_BC.best_params_)
print(grid_result_BC.best_score_)
```

```python
BC = BaggingClassifier(max_features = 0.6, max_samples = 0.8, n_estimators = 200)
BC.fit(Xt_train,Yt_train)
y_predtest = BC.predict(Xt_test)
print(classification_report(Yt_test,y_predtest))
```

```python
k = StratifiedKFold(n_splits = 10)
result_bag = cross_val_score(BC,Xt_train,Yt_train, cv = k)
```

```python
print(result_bag)
print(np.mean(result_bag).round(2))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
GBC = GradientBoostingClassifier()
param_grid ={
    'n_estimators' : [200, 300, 400, 500],
    'max_features' : ['auto', 'sqrt'],
    'max_depth' : [2, 4, 5, 6],
    'min_samples_split' : [2, 5, 10]
}
grid_result  = GridSearchCV(GBC,param_grid,cv = 5,n_jobs = -1)
grid_result_GBC = grid_result.fit(Xt_train,Yt_train)
print(grid_result_GBC.best_params_)
print(grid_result_GBC.best_score_)
```

```python
GBC = GradientBoostingClassifier(max_depth = 6, max_features = "sqrt", min_samples_split = 2,n_estimators = 400)
GBC.fit(Xt_train,Yt_train)
y_predGB = GBC.predict(Xt_test)
print(classification_report(Yt_test, y_predGB))
```

```python
k = StratifiedKFold(n_splits = 10)
result_GB = cross_val_score(GBC,Xt_train,Yt_train,cv = k)
print(result_GB)
print(np.mean(result_GB).round(2))
```
